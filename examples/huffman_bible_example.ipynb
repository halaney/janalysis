```python
import janalysis.huffman as huffman
import pprint
```

```python
# Read in the Bible
bible = 'bible.txt'
text = ''
with open(bible, 'r') as filepointer:
    text = filepointer.read()
```

```python
# Get frequencies of every character
frequencies = {}
for character in list(text):
    if character in frequencies:
        # Increment character count
        frequencies[character] += 1
    else:
        # Add a new character
        frequencies[character] = 1
pprint.pprint(frequencies)
```

```python
# Get huffman codes for these characters
huffman_codes = huffman.get_huffman_codes(list(text))
pprint.pprint(huffman_codes)
```

```python
# Calculate how many MB to store as ascii vs our huffman codes
ascii_bits = 0
huffman_bits = 0
for character, frequency in frequencies.items():
    ascii_bits += frequency * 8  # 8 bits to represent any character as ascii
    huffman_bits += frequency * len(huffman_codes[character])
ascii_megabytes = ascii_bits / (8 * 1000000)
huffman_megabytes = huffman_bits / (8 * 1000000)
print('To store the Bible as an ASCII file: {:.2f} MB'.format(ascii_megabytes))
print('To store the Bible as a huffman code: {:.2f} MB'.format(huffman_megabytes))
print('The huffman Bible is {:.0f}% smaller'.format(100 * (ascii_megabytes - huffman_megabytes)/ascii_megabytes))
```
